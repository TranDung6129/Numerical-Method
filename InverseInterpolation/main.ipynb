# Data 

```python
# Method to input .csv file
def read_csv(filename):
    try:
        df = pd.read_csv(filename, header=None) # Delete header
        # df = pd.read_csv(filename)
        x = df.iloc[:, 0].values
        y = df.iloc[:, 1].values
        return x, y
    except Exception as e:
        print("Error reading CSV file: {e}")
        return None

# Method to input .txt file
def read_txt(filename):
    try:
        df = pd.read_csv(filename, sep='\t', header=None)
        x = df.iloc[:, 0].values
        y = df.iloc[:, 1].values
        return x, y
    except Exception as e:
        print(f"Error reading TXT file: {e}")
        return None
```
